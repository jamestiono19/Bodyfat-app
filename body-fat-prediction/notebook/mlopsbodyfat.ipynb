```python
# ============================================================
# IMPORT LIBRARIES
# ============================================================

from pathlib import Path

import joblib
import mlflow
import mlflow.sklearn
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns

from mlflow.models import infer_signature

from sklearn.ensemble import RandomForestRegressor
from sklearn.linear_model import LinearRegression, Ridge, Lasso
from sklearn.metrics import (
    mean_absolute_error,
    mean_squared_error,
    r2_score
)
from sklearn.model_selection import GridSearchCV, train_test_split
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler


# ============================================================
# PROJECT CONFIGURATION
# ============================================================

PROJECT_DIR = Path(
    r"C:\Users\HP\James T\ML\body-fat-prediction"
)

DATA_PATH = PROJECT_DIR / "data" / "bodyfat.csv"
MODEL_DIR = PROJECT_DIR / "models"
ARTIFACT_DIR = PROJECT_DIR / "artifacts"
EDA_DIR = ARTIFACT_DIR / "eda"
PREDICTION_DIR = ARTIFACT_DIR / "predictions"

MODEL_DIR.mkdir(parents=True, exist_ok=True)
EDA_DIR.mkdir(parents=True, exist_ok=True)
PREDICTION_DIR.mkdir(parents=True, exist_ok=True)


# ============================================================
# MLFLOW CONFIGURATION
# ============================================================

MLFLOW_DB_PATH = PROJECT_DIR / "mlflow.db"

tracking_uri = f"sqlite:///{MLFLOW_DB_PATH.as_posix()}"

mlflow.set_tracking_uri(tracking_uri)
mlflow.set_experiment("BodyFat Prediction")

print("MLflow Tracking URI:", mlflow.get_tracking_uri())


# ============================================================
# LOAD DATASET
# ============================================================

bodyfat = pd.read_csv(DATA_PATH)

print("First five rows:")
print(bodyfat.head())

print("\nDataset information:")
bodyfat.info()

print("\nDescriptive statistics:")
print(bodyfat.describe())

print("\nMissing values:")
print(bodyfat.isna().sum())


# ============================================================
# DATA CLEANING
# ============================================================

original_rows = len(bodyfat)

bodyfat = bodyfat.drop_duplicates().copy()

duplicate_rows_removed = original_rows - len(bodyfat)

print("\nOriginal rows:", original_rows)
print("Duplicate rows removed:", duplicate_rows_removed)
print("Rows after duplicate removal:", len(bodyfat))


# ============================================================
# EXPLORATORY DATA ANALYSIS
# ============================================================

correlation_matrix = bodyfat.corr(numeric_only=True)

# ------------------------------------------------------------
# Correlation heatmap
# ------------------------------------------------------------

plt.figure(figsize=(16, 12))

sns.heatmap(
    correlation_matrix,
    annot=True,
    fmt=".2f",
    cmap="coolwarm"
)

plt.title("Feature Correlation Heatmap")
plt.tight_layout()

heatmap_path = EDA_DIR / "correlation_heatmap.png"
plt.savefig(heatmap_path, dpi=300, bbox_inches="tight")

plt.show()
plt.close()


# ------------------------------------------------------------
# Strongest correlation with BodyFat
# ------------------------------------------------------------

strong_corr = (
    correlation_matrix["BodyFat"]
    .abs()
    .sort_values(ascending=False)
)

print("\nStrongest correlation with BodyFat:")
print(strong_corr)

strong_corr_path = EDA_DIR / "strong_correlations.csv"

strong_corr.rename("AbsoluteCorrelation").to_csv(
    strong_corr_path
)


# ------------------------------------------------------------
# Body fat distribution
# ------------------------------------------------------------

plt.figure(figsize=(8, 5))

sns.histplot(
    bodyfat["BodyFat"],
    kde=True
)

plt.title("Distribution of Body Fat Percentage")
plt.xlabel("Body Fat Percentage")
plt.ylabel("Frequency")
plt.tight_layout()

distribution_path = EDA_DIR / "bodyfat_distribution.png"
plt.savefig(distribution_path, dpi=300, bbox_inches="tight")

plt.show()
plt.close()


# ------------------------------------------------------------
# Abdomen vs BodyFat scatterplot
# ------------------------------------------------------------

plt.figure(figsize=(8, 5))

sns.scatterplot(
    x="Abdomen",
    y="BodyFat",
    data=bodyfat
)

plt.title("Abdomen vs Body Fat")
plt.xlabel("Abdomen Circumference")
plt.ylabel("Body Fat Percentage")
plt.tight_layout()

abdomen_path = EDA_DIR / "abdomen_vs_bodyfat.png"
plt.savefig(abdomen_path, dpi=300, bbox_inches="tight")

plt.show()
plt.close()


# ============================================================
# DEFINE FEATURES AND TARGET
# ============================================================

# Density tidak digunakan karena sangat berkaitan langsung dengan
# perhitungan BodyFat dan berpotensi menyebabkan target leakage.

x = bodyfat.drop(columns=["BodyFat", "Density"])
y = bodyfat["BodyFat"]


# ============================================================
# TRAIN TEST SPLIT
# ============================================================

x_train, x_test, y_train, y_test = train_test_split(
    x,
    y,
    test_size=0.30,
    random_state=42
)

print("\nTraining data shape:", x_train.shape)
print("Testing data shape:", x_test.shape)


# ============================================================
# OUTLIER REMOVAL
# ============================================================

# Outlier threshold dihitung dari training data saja.
# Test data tidak diubah supaya tetap merepresentasikan data baru.

training_data = x_train.copy()
training_data["BodyFat"] = y_train

q1 = training_data.quantile(0.25)
q3 = training_data.quantile(0.75)
iqr = q3 - q1

lower_bound = q1 - (1.5 * iqr)
upper_bound = q3 + (1.5 * iqr)

outlier_mask = (
    (training_data < lower_bound)
    | (training_data > upper_bound)
).any(axis=1)

x_train_clean = x_train.loc[~outlier_mask].copy()
y_train_clean = y_train.loc[~outlier_mask].copy()

outliers_removed = int(outlier_mask.sum())

print("\nTraining outliers removed:", outliers_removed)
print("Clean training data shape:", x_train_clean.shape)


# ============================================================
# LOG DATA PREPARATION AND EDA
# ============================================================

with mlflow.start_run(run_name="EDA_and_Preprocessing"):

    mlflow.set_tags({
        "project": "Body Fat Prediction",
        "run_type": "data_preparation",
        "target": "BodyFat"
    })

    mlflow.log_params({
        "dataset_name": DATA_PATH.name,
        "test_size": 0.30,
        "random_state": 42,
        "outlier_method": "IQR",
        "iqr_multiplier": 1.5,
        "target_column": "BodyFat",
        "excluded_column": "Density"
    })

    mlflow.log_metrics({
        "original_rows": original_rows,
        "duplicate_rows_removed": duplicate_rows_removed,
        "rows_after_duplicates": len(bodyfat),
        "training_rows_before_outliers": len(x_train),
        "training_outliers_removed": outliers_removed,
        "training_rows_after_outliers": len(x_train_clean),
        "testing_rows": len(x_test),
        "number_of_features": x.shape[1]
    })

    # Log seluruh grafik dan CSV EDA
    mlflow.log_artifacts(
        str(EDA_DIR),
        artifact_path="eda"
    )

    # Simpan daftar fitur
    mlflow.log_dict(
        {"features": x.columns.tolist()},
        "metadata/features.json"
    )


# ============================================================
# MODEL AND HYPERPARAMETER CONFIGURATION
# ============================================================

models = {
    "LinearRegression": {
        "estimator": LinearRegression(),
        "params": {}
    },

    "Ridge": {
        "estimator": Ridge(),
        "params": {
            "model__alpha": [0.1, 1.0, 10.0, 100.0]
        }
    },

    "Lasso": {
        "estimator": Lasso(max_iter=10000),
        "params": {
            "model__alpha": [0.001, 0.01, 0.1, 1.0]
        }
    },

    "RandomForest": {
        "estimator": RandomForestRegressor(
            random_state=42
        ),
        "params": {
            "model__n_estimators": [100, 200, 300],
            "model__max_depth": [None, 10, 20, 30],
            "model__min_samples_split": [2, 5, 10],
            "model__min_samples_leaf": [1, 2, 4],
            "model__max_features": ["sqrt", "log2", None]
        }
    }
}


# ============================================================
# TRAIN, TUNE, EVALUATE, AND LOG MODELS
# ============================================================

best_models = {}
model_run_information = {}
results = []

for model_name, configuration in models.items():

    estimator = configuration["estimator"]
    parameter_grid = configuration["params"]

    # Tree-based models tidak memerlukan standardization.
    if model_name == "RandomForest":
        preprocessing_step = "passthrough"
    else:
        preprocessing_step = StandardScaler()

    pipeline = Pipeline([
        ("scaler", preprocessing_step),
        ("model", estimator)
    ])

    grid_search = GridSearchCV(
        estimator=pipeline,
        param_grid=parameter_grid,
        cv=5,
        scoring="r2",
        n_jobs=-1,
        return_train_score=True
    )

    with mlflow.start_run(run_name=model_name) as run:

        mlflow.set_tags({
            "project": "Body Fat Prediction",
            "run_type": "model_training",
            "algorithm": model_name,
            "target": "BodyFat"
        })

        # ----------------------------------------------------
        # Model training and hyperparameter tuning
        # ----------------------------------------------------

        grid_search.fit(
            x_train_clean,
            y_train_clean
        )

        best_pipeline = grid_search.best_estimator_

        best_models[model_name] = best_pipeline

        # ----------------------------------------------------
        # Prediction
        # ----------------------------------------------------

        y_pred = best_pipeline.predict(x_test)

        # ----------------------------------------------------
        # Evaluation
        # ----------------------------------------------------

        mae = mean_absolute_error(
            y_test,
            y_pred
        )

        rmse = np.sqrt(
            mean_squared_error(
                y_test,
                y_pred
            )
        )

        r2 = r2_score(
            y_test,
            y_pred
        )

        best_index = grid_search.best_index_

        cv_r2_mean = grid_search.best_score_

        cv_r2_std = grid_search.cv_results_[
            "std_test_score"
        ][best_index]

        train_r2_mean = grid_search.cv_results_[
            "mean_train_score"
        ][best_index]

        # ----------------------------------------------------
        # Log general parameters
        # ----------------------------------------------------

        mlflow.log_params({
            "model_name": model_name,
            "cv_folds": 5,
            "cv_scoring": "r2",
            "test_size": 0.30,
            "random_state": 42,
            "training_rows": len(x_train_clean),
            "testing_rows": len(x_test),
            "number_of_features": x.shape[1],
            "uses_standard_scaler": model_name != "RandomForest"
        })

        # ----------------------------------------------------
        # Log best hyperparameters
        # ----------------------------------------------------

        if grid_search.best_params_:

            cleaned_parameters = {}

            for parameter_name, parameter_value in (
                grid_search.best_params_.items()
            ):
                clean_name = parameter_name.replace(
                    "model__",
                    ""
                )

                # Mengubah None menjadi string agar aman dicatat.
                if parameter_value is None:
                    parameter_value = "None"

                cleaned_parameters[
                    f"best_{clean_name}"
                ] = parameter_value

            mlflow.log_params(cleaned_parameters)

        # ----------------------------------------------------
        # Log evaluation metrics
        # ----------------------------------------------------

        mlflow.log_metrics({
            "cv_r2_mean": float(cv_r2_mean),
            "cv_r2_std": float(cv_r2_std),
            "cv_train_r2_mean": float(train_r2_mean),
            "test_mae": float(mae),
            "test_rmse": float(rmse),
            "test_r2": float(r2)
        })

        # ----------------------------------------------------
        # Save prediction result
        # ----------------------------------------------------

        prediction_result = pd.DataFrame({
            "ActualBodyFat": y_test.to_numpy(),
            "PredictedBodyFat": y_pred,
            "Residual": y_test.to_numpy() - y_pred
        })

        prediction_path = (
            PREDICTION_DIR
            / f"{model_name}_predictions.csv"
        )

        prediction_result.to_csv(
            prediction_path,
            index=False
        )

        mlflow.log_artifact(
            str(prediction_path),
            artifact_path="predictions"
        )

        # ----------------------------------------------------
        # Residual plot
        # ----------------------------------------------------

        residual = y_test.to_numpy() - y_pred

        residual_figure, residual_axis = plt.subplots(
            figsize=(8, 5)
        )

        residual_axis.scatter(
            y_pred,
            residual,
            alpha=0.7
        )

        residual_axis.axhline(
            y=0,
            linestyle="--"
        )

        residual_axis.set_title(
            f"{model_name} Residual Plot"
        )

        residual_axis.set_xlabel(
            "Predicted Body Fat"
        )

        residual_axis.set_ylabel(
            "Residual"
        )

        residual_figure.tight_layout()

        mlflow.log_figure(
            residual_figure,
            f"plots/{model_name}_residual_plot.png"
        )

        plt.show()
        plt.close(residual_figure)

        # ----------------------------------------------------
        # Model signature and model artifact
        # ----------------------------------------------------

        input_example = x_train_clean.head(3)

        example_prediction = best_pipeline.predict(
            input_example
        )

        signature = infer_signature(
            input_example,
            example_prediction
        )

        model_info = mlflow.sklearn.log_model(
            sk_model=best_pipeline,
            name="model",
            signature=signature,
            input_example=input_example
        )

        # ----------------------------------------------------
        # Store run information
        # ----------------------------------------------------

        model_run_information[model_name] = {
            "run_id": run.info.run_id,
            "model_uri": model_info.model_uri
        }

        results.append({
            "Model": model_name,
            "CV_R2": cv_r2_mean,
            "CV_R2_STD": cv_r2_std,
            "Train_CV_R2": train_r2_mean,
            "Test_MAE": mae,
            "Test_RMSE": rmse,
            "Test_R2": r2,
            "Run_ID": run.info.run_id
        })

        print("\n====================================")
        print("Model:", model_name)
        print("Best parameters:", grid_search.best_params_)
        print(f"CV R² mean : {cv_r2_mean:.4f}")
        print(f"CV R² std  : {cv_r2_std:.4f}")
        print(f"MAE        : {mae:.4f}")
        print(f"RMSE       : {rmse:.4f}")
        print(f"Test R²    : {r2:.4f}")
        print("MLflow run :", run.info.run_id)
        print("====================================")


# ============================================================
# COMPARE MODELS
# ============================================================

results_df = pd.DataFrame(results)

# Champion dipilih menggunakan CV R², bukan test R².
results_df = results_df.sort_values(
    by="CV_R2",
    ascending=False
).reset_index(drop=True)

print("\nModel comparison:")
print(results_df)


# ============================================================
# SELECT CHAMPION MODEL
# ============================================================

champion_name = results_df.loc[0, "Model"]
champion_model = best_models[champion_name]

print("\nChampion model:", champion_name)
print(
    "Champion CV R²:",
    results_df.loc[0, "CV_R2"]
)
print(
    "Champion test R²:",
    results_df.loc[0, "Test_R2"]
)


# ============================================================
# SAVE COMPARISON RESULT
# ============================================================

comparison_path = ARTIFACT_DIR / "model_comparison.csv"

results_df.to_csv(
    comparison_path,
    index=False
)


# ============================================================
# SAVE CHAMPION PIPELINE
# ============================================================

champion_path = MODEL_DIR / "champion_pipeline.pkl"

joblib.dump(
    champion_model,
    champion_path
)

print("\nChampion pipeline saved to:")
print(champion_path)


# ============================================================
# SAVE FEATURE NAMES
# ============================================================

feature_names = x.columns

feature_path = MODEL_DIR / "features.pkl"

joblib.dump(
    feature_names,
    feature_path
)

print("\nFeatures:")
print(feature_names)

print("Feature shape:", x.shape)


# ============================================================
# SAVE LINEAR REGRESSION FOR EXISTING STREAMLIT APPLICATION
# ============================================================

# Bagian ini mempertahankan struktur file lama Anda.
# Streamlit masih dapat menggunakan scaler.pkl dan
# linear_regression_model.pkl secara terpisah.

linear_pipeline = best_models["LinearRegression"]

linear_scaler = linear_pipeline.named_steps["scaler"]
linear_model = linear_pipeline.named_steps["model"]

joblib.dump(
    linear_scaler,
    MODEL_DIR / "scaler.pkl"
)

joblib.dump(
    linear_model,
    MODEL_DIR / "linear_regression_model.pkl"
)

joblib.dump(
    linear_pipeline,
    MODEL_DIR / "linear_regression_pipeline.pkl"
)

print("\nLinear Regression files saved successfully.")


# ============================================================
# SAVE LINEAR REGRESSION COEFFICIENTS
# ============================================================

if hasattr(linear_model, "coef_"):

    coefficients = pd.DataFrame({
        "Feature": x.columns,
        "Coefficient": linear_model.coef_
    })

    coefficients["AbsoluteCoefficient"] = (
        coefficients["Coefficient"].abs()
    )

    coefficients = coefficients.sort_values(
        "AbsoluteCoefficient",
        ascending=False
    )

    coefficient_path = (
        MODEL_DIR / "coefficients.csv"
    )

    coefficients.to_csv(
        coefficient_path,
        index=False
    )

    print("\nFeature importance saved successfully.")
    print(coefficients)


# ============================================================
# LOG FINAL MODEL COMPARISON
# ============================================================

with mlflow.start_run(
    run_name="Final_Model_Comparison"
):

    mlflow.set_tags({
        "project": "Body Fat Prediction",
        "run_type": "model_comparison",
        "champion_model": champion_name
    })

    champion_row = results_df.iloc[0]

    mlflow.log_params({
        "champion_model": champion_name,
        "selection_metric": "CV_R2",
        "number_of_models": len(results_df),
        "champion_run_id": (
            model_run_information[
                champion_name
            ]["run_id"]
        )
    })

    mlflow.log_metrics({
        "champion_cv_r2": float(
            champion_row["CV_R2"]
        ),
        "champion_cv_r2_std": float(
            champion_row["CV_R2_STD"]
        ),
        "champion_test_mae": float(
            champion_row["Test_MAE"]
        ),
        "champion_test_rmse": float(
            champion_row["Test_RMSE"]
        ),
        "champion_test_r2": float(
            champion_row["Test_R2"]
        )
    })

    mlflow.log_artifact(
        str(comparison_path),
        artifact_path="comparison"
    )

    mlflow.log_artifact(
        str(champion_path),
        artifact_path="saved_models"
    )

    mlflow.log_artifact(
        str(feature_path),
        artifact_path="metadata"
    )

    if (
        MODEL_DIR / "coefficients.csv"
    ).exists():

        mlflow.log_artifact(
            str(MODEL_DIR / "coefficients.csv"),
            artifact_path="feature_importance"
        )


print("\nTraining and MLflow logging completed.")
print("Open MLflow UI at http://127.0.0.1:5000")
```
